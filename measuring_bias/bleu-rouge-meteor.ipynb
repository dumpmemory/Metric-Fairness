```python
from nltk.translate.bleu_score import sentence_bleu, corpus_bleu
from nltk.translate.bleu_score import SmoothingFunction

with open("../input/agebias/hyps.txt") as f:
    candidates = [line.strip() for line in f]

with open("../input/agebias/refs.txt") as f:
    references = [line.strip() for line in f]

for i in range(len(candidates)):
    reference = [references[i].split()]
    candidate = candidates[i].split()
    smooth = SmoothingFunction()  # 定义平滑函数对象
    score = sentence_bleu(reference, candidate, weights=(0.25,0.25, 0.25, 0.25), smoothing_function=smooth.method1)
    corpus_score = corpus_bleu([reference], [candidate], smoothing_function=smooth.method1)
    print(score)
```

```python
# !pip install -U nltk
import nltk
nltk.download('omw-1.4')
# from nltk.corpus import wordnet
# nltk.download('wordnet')
with open("../input/genderbias/hyps.txt") as f:
    candidates = [line.strip() for line in f]

with open("../input/genderbias/refs.txt") as f:
    references = [line.strip() for line in f]

    
for i in range(len(candidates)):
    hypothesis = candidates[i].split()
    reference = references[i].split()
    hypothesis=set(hypothesis)
    reference=set(reference)
    #there may be several references
    merteor_score = nltk.translate.meteor_score.single_meteor_score(reference, hypothesis)
    print(merteor_score)
```
