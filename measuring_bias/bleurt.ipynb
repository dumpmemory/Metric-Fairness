```python
from sys import path
path.append(r"../input/bleurt")
from bleurt import score

checkpoint = "../input/bleurt/bleurt-tiny-512"
dataset_list=['race-bias','bertscore','religionbias','physical-bias','agebias','disabilitybias','socioeconomicbias']
scorer = score.BleurtScorer(checkpoint)
with open('./avg.txt', 'w') as f1:
    for path in dataset_list:
   
        with open('../input/'+path + "/hyps.txt") as f:
            candidates = [line.strip() for line in f]

        with open('../input/'+path + "/refs.txt") as f:
            references = [line.strip() for line in f]
    
        scores = scorer.score(references=references, candidates=candidates)

        with open('./' + path+'.txt','w') as f:
            for i in scores:
                    f.write(str(i))
                    f.write('\n')
        data = []
        with open('./' + path+'.txt', 'r') as f:
            lines = f.readlines()
            for line in lines:
                l = float(line.strip())
                data.append(l)
        max_value = max(data)
        min_value = min(data)
        normalized_data = []
        for x in data:
            tmp = (x - min_value) / (max_value - min_value) * 100
            normalized_data.append(tmp)

        import math
        avg_delta = []
        tmp = None
        for x in normalized_data:
            if tmp is None:
                tmp = x
            else:
                tmp = math.fabs(tmp - x)
                avg_delta.append(tmp)
                tmp = None

        f1.write(str(sum(avg_delta) / len(avg_delta)))
        f1.write('\n')
```
